```python
%matplotlib inline
import tsne
import input_data
import numpy as np
import pylab as py
import timeit
```

```python
train, forecast = input_data.get_data('20160830')
num_of_cols = train.shape[1] - 1
X = train.iloc[:,0:num_of_cols]
Y = train.iloc[:,num_of_cols]
train = None
```

```python
no_dims = 3
initial_dims = 20
perplexity = 40.0
max_iter = 200
data_partition = 500
```

```python
start_it = timeit.default_timer()
ret = tsne.tsne(X.iloc[::data_partition,:].as_matrix(), no_dims=no_dims, initial_dims=initial_dims, max_iter=max_iter, perplexity=perplexity)
end_it = timeit.default_timer()
print("Running time : %s" % (end_it - start_it))
```

```python
for i in range(0, no_dims - 1):
    py.scatter(ret[:,i], ret[:,i+1], c=Y[::data_partition].values)
    py.show()
```
